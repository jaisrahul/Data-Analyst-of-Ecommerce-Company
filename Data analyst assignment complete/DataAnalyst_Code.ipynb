```python
import pandas as pd

# Read the data from the Excel files
website_order_report = pd.read_excel(r'C:\Users\LENOVO\Desktop\excelFile\Company X - Order Report.xlsx')
warehouse_pincode_mapping = pd.read_excel(r'C:\Users\LENOVO\Desktop\excelFile\Company X - Pincode Zones.xlsx')
sku_master = pd.read_excel(r'C:\Users\LENOVO\Desktop\excelFile\Company X - SKU Master.xlsx')
courier_company_invoice = pd.read_excel(r'C:\Users\LENOVO\Desktop\excelFile\Courier Company - Invoice.xlsx')
courier_charges_rate_card = pd.read_excel(r'C:\Users\LENOVO\Desktop\excelFile\Courier Company - Rates.xlsx')
Expected_Result = pd.read_excel(r'C:\Users\LENOVO\Desktop\excelFile\Expected_Result.xlsx')

# Merge dataframes to get relevant information
merged_data = website_order_report.merge(warehouse_pincode_mapping, left_on='ExternOrderNo', right_on='Customer Pincode', how='left')
merged_data = merged_data.merge(sku_master, on='SKU', how='left')
merged_data = merged_data.merge(courier_company_invoice, left_on='ExternOrderNo', right_on='Order ID', how='left')

# Drop duplicate values from 'Weight Slabs' column in courier_charges_rate_card dataframe
courier_charges_rate_card = courier_charges_rate_card.drop_duplicates(subset='Weight Slabs')

# Merge with courier_charges_rate_card after dropping duplicates
merged_data = merged_data.merge(courier_charges_rate_card, left_on='Zone_x', right_on='Zone', how='left')

# print(courier_company_invoice.columns)
# print(merged_data)

# Calculate Total weight as per X (KG) and Weight slab as per X (KG)
merged_data['Total weight as per X (KG)'] = merged_data['Order Qty'] * (merged_data['Weight (g)'] / 1000)
merged_data['Weight slab as per X (KG)'] = pd.cut(merged_data['Total weight as per X (KG)'], bins=merged_data['Weight Slabs'], duplicates='drop').apply(lambda x: x.mid)
merged_data['Billing Amount (Rs.)']
# Calculate the Difference Between Expected Charges and Weight slab as per X (Rs.)
merged_data['Difference Between Expected Charges and Billed Charges (Rs.)'] = merged_data['Total weight as per X (KG)'] - merged_data['Billing Amount (Rs.)']

total_orders_correctly_charged = (merged_data['Difference Between Expected Charges and Billed Charges (Rs.)'] == 0).sum()
total_orders_overcharged = (merged_data['Difference Between Expected Charges and Billed Charges (Rs.)'] > 0).sum()
total_orders_undercharged = (merged_data['Difference Between Expected Charges and Billed Charges (Rs.)'] < 0).sum()

total_amount_overcharged = merged_data.loc[merged_data['Difference Between Expected Charges and Billed Charges (Rs.)'] > 0, 'Difference Between Expected Charges and Billed Charges (Rs.)'].sum()
total_amount_undercharged = merged_data.loc[merged_data['Difference Between Expected Charges and Billed Charges (Rs.)'] < 0, 'Difference Between Expected Charges and Billed Charges (Rs.)'].sum()

# Create the summary table DataFrame
summary_table_data = {
    'Category': ['Total orders where X has been correctly charged', 'Total Orders where X has been overcharged', 'Total Orders where X has been undercharged'],
    'Count': [total_orders_correctly_charged, total_orders_overcharged, total_orders_undercharged],
    'Amount (Rs.)': [0, total_amount_overcharged, total_amount_undercharged]
}
summary_table_df = pd.DataFrame(summary_table_data)

summary_table_df.to_csv(r'C:\Users\LENOVO\Desktop\Data analyst assignment complete\summary_table.csv', index=False)

summary_table_df
```
